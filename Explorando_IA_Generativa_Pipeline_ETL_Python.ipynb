# Etapa de Load
Carregar dados do arquivo Pipeline_ETL_Python.csv para anÃ¡lise e tratativa dos dados.


```python
import pandas as pd

def read_csv_robust(filepath, encoding='latin1'):
    with open(filepath, 'r', encoding=encoding) as f:
        lines = f.readlines()

    # Extract header from the first line, stripping quotes
    header = [h.strip().strip('"') for h in lines[0].strip().split(',')]

    # Process data lines, stripping quotes
    data = []
    for line in lines[1:]:
        if line.strip(): # Avoid processing empty lines
            data.append([d.strip().strip('"') for d in line.strip().split(',')])

    return pd.DataFrame(data, columns=header)

# LÃª o CSV usando a funÃ§Ã£o robusta e converte para uma lista de dicionÃ¡rios
df_users = read_csv_robust('Pipeline_ETL_Python.csv', encoding='latin1')
users = df_users.to_dict(orient='records')

# Garante a estrutura esperada para a etapa de TransformaÃ§Ã£o
for user in users:
    # Initialize 'news' as an empty list, and ensure 'id' is an integer
    user['news'] = []
    if 'UserID' in user: # Using 'UserID' as per CSV, convert to 'id' if needed later for API
        user['id'] = int(user.pop('UserID')) # Convert and rename for consistency with API structure
```

```python
import openai

# Initialize the OpenAI client with your API key
client = openai.OpenAI(api_key=openai_api_key)

def generate_ai_news(user):
  completion = client.chat.completions.create(
    model="gpt-4",
    messages=[
      {
          "role": "system",
          "content": "VocÃª Ã© um especialista em markting bancÃ¡rio."
      },
      {
          "role": "user",
          "content": f"Crie uma mensagem para {user['Name']} sobre a importÃ¢ncia dos investimentos (mÃ¡ximo de 100 caracteres)"
      }
    ]
  )
  # Accessing the content from the new response object structure
  return completion.choices[0].message.content.strip('"')

# Process only the first 5 users
for i in range(min(5, len(users))):
    user = users[i]
    news = generate_ai_news(user)
    print(f"Mensagem para {user['Name']}: {news}")
    user['news'].append({
        "icon": "https://digitalinnovationone.github.io/santander-dev-week-2023-api/icons/credit.svg",
        "description": news
    })
```
